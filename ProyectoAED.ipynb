# Minnesota's DataFrame

```python
import pandas as pd
# Se crea el dataframe de minnesota desde el CSV en el repositorio de github
mn_df = pd.read_csv("https://raw.githubusercontent.com/valentinabc19/Proyecto1-Estadistica1/main/samples/minnesota_sample.csv")

# Se seleccionan solo las columnas de interés
mn_df = mn_df[['Region','Gender','Handed', 'Favourite_physical_activity', 'Importance_reducing_pollution',
               'Birth_month', 'Beverage', 'Favorite_School_Subject', 'Sleep_Hours_Non_Schoolnight',
               'Paid_Work_Hours', 'Work_At_Home_Hours','Planned_Education_Level']]

# Mostrar el DataFrame
mn_df
len(mn_df) #Verificación de total de filas
```

```python
mn_df = mn_df.dropna() #Se eliminan las filas con valores nulos
len(mn_df) #Verificación de total de filas restantes
#mn_df
```

```python
# Convierte la columna 'Importance_reducing_pollution' a numérico y maneja los valores no numéricos de manera adecuada
mn_df['Importance_reducing_pollution'] = pd.to_numeric(mn_df['Importance_reducing_pollution'], errors='coerce').astype(int)

# Elimina los valores que no están dentro del rango válido (0 a 1000)
mn_df = mn_df[(mn_df['Importance_reducing_pollution'] >= 0) & (mn_df['Importance_reducing_pollution'] <= 1000)]

mn_df[['Importance_reducing_pollution']]
len(mn_df)
```

```python
"""
Se limpia específicamente la columna "Sleep_Hours_Non_Schoolnight" ya que los datos que esta contiene
son digitados manualmente por los usuarios, por lo que es más susceptible a tener valores indeseados
"""
import numpy as np

# Función para limpiar la columna 'Sleep_Hours_Non_Schoolnight'
def clean_sleep_hours(value):
    try:
        # Elimina cualquier carácter no numérico y espacios en blanco
        value = value.strip().lower().replace('hrs', '').replace(' ', '')

        if '-' in value:  # Convierte intervalos en sus promedios
            start, end = map(float, value.split('-'))
            return round((start + end) / 2)
        elif value.isdigit() or value.replace('.', '', 1).isdigit():  # Maneja números y decimales
            value = float(value)
            if 0 <= value <= 24:  # Considera valores dentro de un rango realista para horas de sueño
                return int(round(value))
        return np.nan  # Asigna NaN a valores inválidos o atípicos
    except:
        return np.nan  # Maneja cualquier caso inesperado

# Aplica la función de limpieza a la columna
mn_df['Sleep_Hours_Non_Schoolnight'] = mn_df['Sleep_Hours_Non_Schoolnight'].apply(lambda x: clean_sleep_hours(str(x)))

# Elimina las filas con NaN en 'Sleep_Hours_Non_Schoolnight' (valores atípicos y entradas inválidas)
mn_df = mn_df.dropna(subset=['Sleep_Hours_Non_Schoolnight'])

# Convierte la columna 'Sleep_Hours_Non_Schoolnight' a tipo entero
mn_df['Sleep_Hours_Non_Schoolnight'] = mn_df['Sleep_Hours_Non_Schoolnight'].astype(int)

# Muestra el dataframe limpio
mn_df[['Sleep_Hours_Non_Schoolnight']]
mn_df
```

```python
# Función para limpiar las columnas en fomrato de horas semanales
def clean_weekly_hours(value):
    try:
        # Elimina cualquier carácter no numérico y espacios en blanco
        value = value.strip().lower().replace('hrs', '').replace(' ', '')

        if '-' in value:  # Convierte intervalos en sus promedios
            start, end = map(float, value.split('-'))
            return round((start + end) / 2)
        elif value.isdigit() or value.replace('.', '', 1).isdigit():  # Maneja números y decimales
            value = float(value)
            if 0 <= value <= 168:  # Considera valores dentro de un rango realista para horas de sueño
                return int(round(value))
        return np.nan  # Asigna NaN a valores inválidos o atípicos
    except:
        return np.nan  # Maneja cualquier caso inesperado

# Aplica la misma función de limpieza a la columna 'Paid_Work_Hours' en los datos de Minesota
mn_df['Paid_Work_Hours'] = mn_df['Paid_Work_Hours'].apply(lambda x: clean_weekly_hours(str(x)))

# Elimina las filas con NaN en 'Paid_Work_Hours' (valores atípicos y entradas inválidas)
mn_df = mn_df.dropna(subset=['Paid_Work_Hours'])

# Convierte la columna 'Paid_Work_Hours' a tipo entero
mn_df['Paid_Work_Hours'] = mn_df['Paid_Work_Hours'].astype(int)

# Muestra el dataframe limpio
mn_df[['Paid_Work_Hours']]
#len(ne_df)
mn_df
```

```python
# Aplica la misma función de limpieza a la columna 'Work_At_Home_Hours' en los datos de Minesota
mn_df['Work_At_Home_Hours'] = mn_df['Work_At_Home_Hours'].apply(lambda x: clean_weekly_hours(str(x)))

# Elimina las filas con NaN en 'Paid_Work_Hours' (valores atípicos y entradas inválidas)
mn_df = mn_df.dropna(subset=['Work_At_Home_Hours'])

# Convierte la columna 'Paid_Work_Hours' a tipo entero
mn_df['Work_At_Home_Hours'] = mn_df['Work_At_Home_Hours'].astype(int)

# Muestra el dataframe limpio
mn_df[['Work_At_Home_Hours']]
#len(ne_df)
mn_df
```

## 4. ¿Cuál es el deporte / actividad favorita de los estudiantes que participan en el Censo escolar en cada estado?

```python
# Encuentra todas las actividades favoritas con el conteo máximo en cada estado
def get_favorite_activity(group):
    counts = group.value_counts()
    max_count = counts.max()
    favorites = counts[counts == max_count].index.tolist()
    return favorites

favorite_activities_by_state = df.groupby('Region')['Favourite_physical_activity'].agg(get_favorite_activity)

# Mostrar el resultado
print(favorite_activities_by_state)
```

### h) ¿Qué tipo de asimetría presenta esta variable?

## 7. Construya una Ojiva para la variable “Tiempo que dedican los estudiantes para ayudar a los quehaceres del Hogar” y a partir de esta gráfica responda las siguientes preguntas:

```python
import plotly.graph_objects as go

# Ordenar los datos por la columna de interés
df_sorted = df.sort_values(by='Work_At_Home_Hours')

# Definir el número de bins
bins = 20

# Calcular los bins manualmente y obtener los límites
df_sorted['Bins'] = pd.cut(df_sorted['Work_At_Home_Hours'], bins=bins)
bins_limits = pd.cut(df_sorted['Work_At_Home_Hours'], bins=bins, retbins=True)[1]

# Calcular la frecuencia acumulada por cada bin y convertirla a enteros
frecuencia_acumulada = df_sorted['Bins'].value_counts().sort_index().cumsum().astype(int)

# Obtener el punto medio de cada bin
bins_medio = 0.5 * (bins_limits[1:] + bins_limits[:-1])

# Crear la ojiva usando Plotly
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=bins_medio,
    y=frecuencia_acumulada,
    mode='lines+markers',
    name='Frecuencia Acumulada'
))

fig.update_layout(
    title='Ojiva de "Tiempo dedicado a los quehaceres del Hogar"',
    xaxis_title='Horas dedicadas',
    yaxis_title='Frecuencia acumulada (número de estudiantes)',
    showlegend=True
)

fig.show()
```
